```python
# Look up the word 'mostrous' in 'Moby Dick'
text1.concordance('monstrous')
```

Using the concordance function puts the search word into context, with some of the words preceding and following it.

We can also perform a search for similar words based on context:

```python
text1.similar('monstrous')
```

```python
text2.similar('monstrous')
```

An interesting observation can be drawn: the authors of these two texts use the word 'mostrous' in very different ways!

Let's look at contexts shared by two words:

```python
text2.common_contexts(['monstrous', 'very'])
```
